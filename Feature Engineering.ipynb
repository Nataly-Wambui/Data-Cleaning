```python
def reduce_cardinality(diabetes_diabetic_df, column, threshold=0.01):
    freq = diabetes_diabetic_df[column].value_counts(normalize=True)
    categories_to_keep = freq[freq > threshold].index.tolist()
    diabetes_diabetic_df[column] = diabetes_diabetic_df[column].apply(lambda x: x if x in categories_to_keep else 'Other')
```

```python
reduce_cardinality(diabetes_diabetic_df, 'payer_code', threshold=0.01)
```

```python
print(diabetes_diabetic_df['payer_code'].value_counts())
```

```python
reduce_cardinality(diabetes_diabetic_df, 'medical_specialty', threshold=0.01)
```

```python
print(diabetes_diabetic_df['medical_specialty'].value_counts())
```

```python
def combine_categories(diabetes_diabetic_df, column, categories_to_combine, new_category_name):
    diabetes_diabetic_df.loc[diabetes_diabetic_df[column].isin(categories_to_combine), column] = new_category_name
    return diabetes_diabetic_df
```

```python
diabetes_diabetic_df = combine_categories(diabetes_diabetic_df, 'diag_1', ['diag_2', 'diag_3'], 'diagnosis')
```
